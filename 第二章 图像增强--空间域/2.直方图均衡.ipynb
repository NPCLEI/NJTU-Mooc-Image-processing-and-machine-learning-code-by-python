```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
plt.figure(1,figsize=(20,20))

def histMean(raw):
    h,w = raw.shape
    #+1
    Dm = np.max(raw) + 1

    dct = [0 for i in range(Dm)]
    #
    A0 = h*w
    #Dm/A0
    C = Dm / A0
    raw = raw.reshape((1,A0))

    #
    for i in range(A0):
        dct[raw[0,i]]+=1

    #
    for i in range(Dm):
        dct[i] = round(C * dct[i])

    #
    for i in range(A0):
        raw[0,i] = dct[raw[0,i]]

    raw = raw.reshape((h,w))

    return raw

raw=cv2.imread('../imgs/c2/cells.png',0)

plt.subplot(1,4,1)
plt.imshow(raw,cmap='gray')
plt.subplot(1,4,2)
plt.hist(raw)

plt.subplot(1,4,3)
plt.imshow(histMean(raw),cmap='gray')
plt.subplot(1,4,4)
plt.hist(raw)
plt.show()
```
